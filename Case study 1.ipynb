```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### 1. Read the dataset to the python environment.

```python
df = pd.read_csv('cars_data.csv')
df
```

### 2. Check for the null values present in the dataset.

```python
df.isnull().sum()
```

### 3. Plot a bar graph of male vs female buyers participated in the sales.

-> total of 5052 Female buyers and 4948 male buyers

```python
plt.figure(figsize = (15,8))
mvf_plot = sns.countplot(data = df, x = 'Buyer Gender', hue = 'Buyer Gender')
for each in mvf_plot.containers:
    mvf_plot.bar_label(each)

plt.show()
```

### 4. Find the top 5 cars based on their sales price

1) Volkswagen Jetta III 2) Audi A6 A6 3) LX 4) MX-6 5) ES

```python
top5_sales = df.sort_values('Sale Price', ascending = False).head()
top5_sales
```

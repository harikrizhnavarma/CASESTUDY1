```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_excel('Telco-Customer-Churn.xlsx')
data
```

### Compare churn count with resp to gender

```python
data_churn = sns.countplot(data = data, x = 'gender', hue = 'Churn')
plt.title('churn of men and women', fontsize = 17)
for each in data_churn.containers:
    data_churn.bar_label(each)
plt.show()
```

### Number of female citizens

**Number of Female citizens - 568**

```python
# All senior citizens
senior_citizen = data.loc[data['SeniorCitizen'] == 1.0]
senior_citizen
```

```python
female_sen_citizen = senior_citizen.loc[senior_citizen['gender'] == 'Female']
print('Number of Female citizens -',female_sen_citizen['gender'].count())
```

### Compare 'tenure' with 'Total Charges'

```python
plt.figure(figsize = (15,8))
plt.hist(data['tenure'], color = 'red', cumulative = True, rwidth = 0.9)
plt.title('tenure vs Total Charges', fontsize = 15)
plt.xlabel('Tenure')
plt.ylabel('Total Charges')
plt.show()
```

### Find out which contract is preferred by the senior citizen.

**Answer - Month to Month Contract**

```python
pref_con = sns.countplot(data = senior_citizen, x = 'Contract')
plt.title('Most preffered contract by senior citizens', fontsize = 15)
for each in pref_con.containers:
    pref_con.bar_label(each)
plt.show()
```
